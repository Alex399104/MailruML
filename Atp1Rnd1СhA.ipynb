```
import numpy as np
from sklearn.model_selection import GridSearchCV
from numpy import genfromtxt
import pandas as pd

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.decomposition import PCA
```

```
DataFrame_Train=pd.read_csv('train.csv', header=None)
DataFrame_Tt=pd.read_csv('train-target.csv', header=None)
DataFrame_Sample=genfromtxt('sample-submission.csv')
DataFrame_Test=pd.read_csv('test.csv', header=None)
DataFrame_Train
```

```
inputs = keras.Input(shape=(30,))

layer_1 = layers.Dense(30,kernel_initializer='random_normal',activation="relu")(inputs)
drop_1 = layers.Dropout(0.25)(layer_1)

layer_2 = layers.Dense(30,kernel_initializer='random_normal',activation="relu")(drop_1)
drop_2 = layers.Dropout(0.25)(layer_2)

layer_3 = layers.Dense(30,kernel_initializer='random_normal',activation="relu")(drop_2)
drop_3 = layers.Dropout(0.25)(layer_3)

softmax_1 = layers.Softmax()(drop_3)
outputs = layers.Dense(1)(softmax_1)
model = keras.Model(inputs=inputs, outputs=outputs, name="Neural_2")

model.compile(loss='mse',optimizer='adam')

model.fit(DataFrame_Train,DataFrame_Tt, batch_size=200, epochs=15, validation_split=0.2)

y_predict=model.predict(DataFrame_Test)

MyData = pd.DataFrame(y_predict)
MyData.to_csv("Machine_Learning_Neural.csv",header=None, index=None)

```
